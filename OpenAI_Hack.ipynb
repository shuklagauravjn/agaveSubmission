```python
import pandas as pd
from pandasai import SmartDataframe
from pandasai.llm import AzureOpenAI
import os
```

```python
df = pd.read_csv('Cleansed_USA_Housing.csv')
```

```python
os.environ["OPENAI_API_TYPE"]="azure"
os.environ["OPENAI_API_VERSION"]="2023-07-01-preview"
os.environ["OPENAI_API_BASE"] = os.environ.get('OPENAI_DEPLOYMENT_ENDPOINT')  
os.environ["OPENAI_API_KEY"] = os.environ.get('OPENAI_API_KEY')
```

```python
dfllm = AzureOpenAI(api_token=os.environ['OPENAI_API_KEY'],
                api_base=os.environ['OPENAI_API_BASE'],
                api_version=os.environ["OPENAI_API_VERSION"],
                deployment_name="gpt-4",
                temperature=0,
                api_type=os.environ['OPENAI_API_TYPE'],
                is_chat_model=True)
```

```python
df = SmartDataframe(df, config={"llm": dfllm})
```

```python
response = df.chat("How many records are there in my dataset?")
print(response)
```
